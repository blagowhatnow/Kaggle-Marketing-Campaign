```python
#Kaggle time series https://www.kaggle.com/datasets/sumanthvrao/daily-climate-time-series-data

#Inspired from https://machinelearningmastery.com/time-series-forecasting/

#Compare with rolling time series using linear regression



import pandas as pd
from pandas import datetime
from matplotlib import pyplot
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
series_train = pd.read_csv('DailyDelhiClimateTrain.csv')
series_test=pd.read_csv('DailyDelhiClimateTest.csv')
```

```python
temp_train=series_train['meantemp']
humidity_train=series_train['humidity']
wind_speed_train=series_train['wind_speed']
meanpressure_train=series_train['meanpressure']

temp_test=series_test['meantemp']
humidity_test=series_test['humidity']
wind_speed_test=series_test['wind_speed']
meanpressure_test=series_test['meanpressure']
```

```python
#Rolling window forecast for temperature on the test set

history=list(temp_train)
predictions=[]

for t in range(len(temp_test)):
    model = ARIMA(history,order=(5,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    predictions.append(output)
    obs = temp_test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (output, obs))
# evaluate forecasts
rmse = sqrt(mean_squared_error(temp_test, predictions))
print('Test RMSE: %.3f' % rmse)
# plot forecasts against actual outcomes
pyplot.plot(temp_test)
pyplot.plot(predictions, color='red')
pyplot.show()
```
