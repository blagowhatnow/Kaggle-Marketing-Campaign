```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
```

```python
df_train=pd.read_csv("train.csv")
df_train.head()
```

```python
df_test=pd.read_csv("test.csv")
df_test.tail()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for (columnName) in df_train.select_dtypes(exclude=["number"]):
    le.fit(df_train[columnName])
    df_train[columnName]=le.transform(df_train[columnName])
    
for (columnName) in df_test.select_dtypes(exclude=["number"]):
    le.fit(df_test[columnName])
    df_test[columnName]=le.transform(df_test[columnName])  
    
df_test=df_test.dropna()
df_train=df_train.dropna()
```

```python
y_train=df_train['SalePrice']
df_train=df_train.drop(['SalePrice'],axis=1)
df_train=df_train.drop(['Id'],axis=1)

X_train=df_train.to_numpy()

df_test=df_test.drop(['Id'],axis=1)

X_test=df_test.to_numpy()
```

```python
from keras.layers import Dense, Activation
from keras.models import Sequential

model = Sequential()

model.add(Dense(128, kernel_initializer='normal',input_shape = X_train[0].shape, activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))

model.add(Dense(1, kernel_initializer='normal',activation='linear'))

model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
model.summary()
```

```python
# now we "compile" the model specifying the loss and optimizer
model.compile(
    loss='mean_squared_error', # this is our cross-entropy
    optimizer='adam',
    metrics=['accuracy']  # report accuracy during training
)
```

```python
model.fit(
    X_train, 
    y_train,
    batch_size=512, 
    epochs=4000,verbose=1)
```

```python
pred=model.predict(X_train)
y_train_pred=[pred[i][0] for i in np.arange(0,X_train.shape[0])]
```

```python
import seaborn as sns
from scipy import stats
def r2(x, y):
    return stats.pearsonr(x, y)[0] ** 2

sns.jointplot(x=y_train_pred,
            y=y_train, 
            kind="reg", 
            scatter_kws={"s": 1});
```

```python
r2(y_train,y_train_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor

reg=RandomForestRegressor()
reg.fit(X_train,y_train)
y_train_pred=reg.predict(X_train)
```

```python
import seaborn as sns
from scipy import stats
def r2(x, y):
    return stats.pearsonr(x, y)[0] ** 2

sns.jointplot(x=y_train_pred,
            y=y_train, 
            kind="reg", 
            scatter_kws={"s": 1});
```

```python
r2(y_train,y_train_pred)
```
